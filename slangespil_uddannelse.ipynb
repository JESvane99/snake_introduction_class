### Nu beregner vi nogle variabler til brug i spillets opsætning
- maksimum- og minimumværdier for `x` og `y` koordinater til at definere grænserne for spilleområdet baseret på den definerede spilbredde og -højde
- `point_x` og `point_y` til at definere positionen for pointvisningen baseret på den definerede spilhøjde og pointskriftstørrelse

> Dette skal afspejle de forskellige matematiske operatorer for at vise forskellen, og hvordan de kan ændre variabeltypen og dermed muligvis output af programmet.

**Matematiske operatorer:**
- `+` plus - lægger sammen to tal
- `-` minus - trækker et tal fra et andet (eller laver et negativt tal)
- `*` gange - multiplicerer to tal
- `/` division - deler et tal med et andet (giver altid et decimaltal/float)
- `//` heltalsdivision - deler og runder ned til nærmeste heltal

```python
# Beregninger med forskellige matematiske operatorer:
x_min = -spil_bredde // 2  # // heltalsdivision for at sikre, at vi får et heltal
x_max = -x_min  # - minus for at gøre det positivt, da x_min er negativ
y_min = int(-spil_højde / 2)  # / almindelig division og int() for at sikre integer
y_max = y_min * -1  # *-1 for at gøre det positivt, da y_min er negativ
point_x = 0
point_y = spil_højde // 2 - (point_skrifttype[1] + 10)  # paranteser respekteres

# Flere eksempler på matematiske operatorer:
samlet_areal = spil_bredde * spil_højde  # * gange
halv_bredde = spil_bredde / 2  # / almindelig division (giver float)
dobbelt_forsinkelse = forsinkelse + forsinkelse  # + plus for at lægge sammen
margin = 10 + 5  # + plus
```

### Derefter kommer funktionsdefinitionerne, som inkluderer:
- `flyt()` til at kontrollere slangens bevægelse baseret på den aktuelle retning
- `gå_op()`, `gå_ned()`, `gå_venstre()`, og `gå_højre()` til at ændre retningen af slangens bevægelse når den tilsvarende tast trykkes

```python
def flyt(hoved, retning):
    """Flyt slangen baseret på den aktuelt indstillede retning."""
    if hoved.retning == "op":
        y = hoved.ycor()
        hoved.sety(y + 20)
    elif hoved.retning == "ned":
        y = hoved.ycor()
        hoved.sety(y - 20)
    elif hoved.retning == "venstre":
        x = hoved.xcor()
        hoved.setx(x - 20)
    elif hoved.retning == "højre":
        x = hoved.xcor()
        hoved.setx(x + 20)
    else:
        pass

def gå_op():
    """skift retning af slangehoved til op hvis det ikke i øjeblikket bevæger sig ned."""
    global hoved
    if hoved.retning != "ned":
        hoved.retning = "op"


def gå_ned():
    """skift retning af slangehoved til ned hvis det ikke i øjeblikket bevæger sig op."""
    global hoved
    if hoved.retning != "op":
        hoved.retning = "ned"


def gå_venstre():
    """skift retning af slangehoved til venstre hvis det ikke i øjeblikket bevæger sig til højre."""
    global hoved
    if hoved.retning != "højre":
        hoved.retning = "venstre"


def gå_højre():
    """skift retning af slangehoved til højre hvis det ikke i øjeblikket bevæger sig til venstre."""
    global hoved
    if hoved.retning != "venstre":
        hoved.retning = "højre"
```

### Turtle-grafikopsætning og spilløkke:

Inkluderer opsætningen af turtle-objekterne til selve spilvinduet, slangens hoved, maden og scorevisningen. Den inkluderer også hovedspilløkken, hvor spillogikken udføres, såsom kontrol for kollisioner, opdatering af scoren og flytning af slangen.
Variablerne og funktionerne defineret i de tidligere sektioner bruges i hele denne del til at kontrollere spilforløbet og opdatere spilstaten baseret på brugerinput og spilhændelser.

```python
# Opsætning af skærmen
vindue = turtle.Screen()
vindue.title(skærm_titel)
vindue.bgcolor(baggrunds_farve)
vindue.setup(width=vindue_bredde, height=vindue_højde)
vindue.tracer(0)  # Slår skærmopdateringerne fra
```

```python
# Slangehoved
hoved = turtle.Turtle()
hoved.speed(0)
hoved.shape(slange_hoved_form)
hoved.color(slange_farve)
hoved.penup()
hoved.goto(0, 0)
hoved.retning = "stop"  # tilføj retningsattribut til hoved turtle-objekt - dette kan gøres på grund af pythons dynamiske typning
slange = [hoved]
```

```python
mad = turtle.Turtle()
mad.speed(0)
mad.shape(mad_form)
mad.color(mad_farve)
mad.penup()
mad.goto(0, 100)
```

```python
# Pen til pointvisning (score display)
pen = turtle.Turtle()
pen.speed(0)
pen.shape(
    "square"
)  # Man kunne også definere en variabel for penformen og bruge den her for at gøre det mere tilpasseligt
pen.color(point_farve)
pen.penup()
pen.hideturtle()
pen.goto(point_x, point_y)
pen.write(f"Point: {point}  Bedste: {bedste_point}", align="center", font=point_skrifttype)
```

```python
# Tastatur-bindinger (Key bindings)
vindue.listen()
vindue.onkeypress(gå_op, knap_op)
vindue.onkeypress(gå_ned, knap_ned)
vindue.onkeypress(gå_venstre, knap_venstre)
vindue.onkeypress(gå_højre, knap_højre)
```

```python
# Hoved spilløkke (Game loop)
while True:
    vindue.update()

    # Tjek for kollision med grænsen
    grænse_kollision = (
        hoved.xcor() > x_max
        or hoved.xcor() < x_min
        or hoved.ycor() > y_max
        or hoved.ycor() < y_min
    )
    slange_kollision = any(
        segment.distance(hoved) < 20 for segment in slange[1:]
    )  # tjek om hoved kolliderer med noget segment i slangens krop (eksklusive hovedet selv)
    if grænse_kollision or slange_kollision:
        hoved.retning = "stop"
        for segment in slange:
            segment.shape("turtle")
            vindue.update()  # opdater skærmen for at animere slangens død.
            time.sleep(forsinkelse)
        [segment.reset() for segment in slange[1:]]

        # Ryd segmentlisten
        hoved.goto(0, 0)
        hoved.shape(slange_hoved_form)  # nulstil hovedformen
        slange = [hoved]

        # Nulstil point
        point = 0

        pen.clear()
        pen.write(
            f"Point: {point}  Bedste: {bedste_point}",
            align="center",
            font=point_skrifttype,
        )

    # Tjek for kollision med maden
    if hoved.distance(mad) < 20:
        # Tilføj et segment
        nyt_segment = turtle.Turtle()
        nyt_segment.speed(0)
        nyt_segment.shape(slange_segment_form)
        nyt_segment.color(slange_farve)
        nyt_segment.penup()
        nyt_segment.goto(mad.xcor(), mad.ycor())
        slange.append(nyt_segment)

        # Flyt maden til et tilfældigt sted
        x = randint(
            x_min, x_max
        )  # brug definerede variabler for min og max x og y for at gøre det mere tilpasseligt
        y = randint(y_min, y_max)
        mad.goto(x, y)

        # Forkort forsinkelsen
        # if øg_forsinkelse:
        #   forsinkelse -= 0.001

        # Øg point
        point += 10

        if point > bedste_point:
            bedste_point = point
            pen.clear()
            pen.write(
                f"Point: {point}  Bedste: {bedste_point}",
                align="center",
                font=point_skrifttype,
            )

    # Flyt endesegmenterne først i omvendt rækkefølge
    for i in range(len(slange) - 1, 0, -1):
        if i > 0:
            x = slange[i - 1].xcor()
            y = slange[i - 1].ycor()
        slange[i].goto(x, y)

    flyt(hoved, hoved.retning)

    time.sleep(forsinkelse)
```
