## Import af nødvendige biblioteker

```python
import turtle
import time
import random
```

## Definition af konstanter og variabler

```python
forsinkelse = 0.1
point = 0
bedste_point = 0
spil_bredde = 600
spil_højde = 600
vindue_bredde = 700
vindue_højde = 700
```

### Bevægelse

```python
knap_op, knap_ned, knap_venstre, knap_højre = "Up", "Down", "Left", "Right"
retning = "stop"
```

### Design

```python
skærm_titel = "Slangespil"
slange_farve = "green"
slange_hoved_form = "square"
slange_segment_form = "square"
mad_farve = "red"
mad_form = "circle"
baggrunds_farve = "black"
point_farve = "white"
point_skrifttype = ("Arial", 24, "normal")
```

### Beregninger med matematiske operatorer

```python
x_min = -spil_bredde // 2
x_max = spil_bredde // 2
y_min = -spil_højde // 2
y_max = spil_højde // 2
point_x = 0
point_y = spil_højde // 2 - point_skrifttype[1] - 10 

samlet_areal = spil_bredde * spil_højde
halv_bredde = spil_bredde / 2
dobbelt_forsinkelse = forsinkelse + forsinkelse
margin = 10 + 5 
```

## bevægelse funktions definitioner

```python
def flyt(hoved, retning):
    if retning == "op":
        y = hoved.ycor()
        hoved.sety(y + 20)
    if retning == "ned":
        y = hoved.ycor()
        hoved.sety(y - 20)
    if retning == "venstre":
        x = hoved.xcor()
        hoved.setx(x - 20)
    if retning == "højre":
        x = hoved.xcor()
        hoved.setx(x + 20)


def gå_op():
    global retning
    if retning != "ned":
        retning = "op"


def gå_ned():
    global retning
    if retning != "op":
        retning = "ned"


def gå_venstre():
    global retning
    if retning != "højre":
        retning = "venstre"


def gå_højre():
    global retning
    if retning != "venstre":
        retning = "højre"
```

## Spil opsætning med turtle biblioteket

```python
vindue = turtle.Screen()
vindue.title(skærm_titel)
vindue.bgcolor(baggrunds_farve)
vindue.setup(width=vindue_bredde, height=vindue_højde)
vindue.tracer(0)
```

```python
hoved = turtle.Turtle()
hoved.speed(0)
hoved.shape(slange_hoved_form)
hoved.color(slange_farve)
hoved.penup()
hoved.goto(0, 0)
retning = "stop"
slange = [hoved]
```

```python
mad = turtle.Turtle()
mad.speed(0)
mad.shape(mad_form)
mad.color(mad_farve)
mad.penup()
mad.goto(0, 100)
```

```python
pen = turtle.Turtle()
pen.speed(0)
pen.shape("square")
pen.color(point_farve)
pen.penup()
pen.hideturtle()
pen.goto(point_x, point_y)
pen.write(f"Point: {point}  Bedste: {bedste_point}", align="center", font=point_skrifttype)
```

```python
vindue.listen()
vindue.onkeypress(gå_op, knap_op)
vindue.onkeypress(gå_ned, knap_ned)
vindue.onkeypress(gå_venstre, knap_venstre)
vindue.onkeypress(gå_højre, knap_højre)
```

## Spil loop

```python
while True:
    vindue.update()

    grænse_kollision = (
        hoved.xcor() > x_max
        or hoved.xcor() < x_min
        or hoved.ycor() > y_max
        or hoved.ycor() < y_min
    )
    slange_kollision = any(
        segment.distance(hoved) < 20 for segment in slange[1:]
    )
    if grænse_kollision or slange_kollision:
        retning = "stop"
        for segment in slange:
            segment.shape("turtle")
            vindue.update()
            time.sleep(forsinkelse)
        [segment.reset() for segment in slange[1:]]

        hoved.goto(0, 0)
        hoved.shape(slange_hoved_form)
        slange = [hoved]

        point = 0

        pen.clear()
        pen.write(
            f"Point: {point}  Bedste: {bedste_point}",
            align="center",
            font=point_skrifttype,
        )

    if hoved.distance(mad) < 20:
        nyt_segment = turtle.Turtle()
        nyt_segment.speed(0)
        nyt_segment.shape(slange_segment_form)
        nyt_segment.color(slange_farve)
        nyt_segment.penup()
        nyt_segment.goto(mad.xcor(), mad.ycor())
        slange.append(nyt_segment)

        x = random.randint(x_min, x_max)
        y = random.randint(y_min, y_max)
        mad.goto(x, y)

        point += 10

        if point > bedste_point:
            bedste_point = point
            pen.clear()
            pen.write(
                f"Point: {point}  Bedste: {bedste_point}",
                align="center",
                font=point_skrifttype,
            )

    for i in range(len(slange) - 1, 0, -1):
        if i > 0:
            x = slange[i - 1].xcor()
            y = slange[i - 1].ycor()
        slange[i].goto(x, y)

    flyt(hoved, retning)

    time.sleep(forsinkelse)
```
