# Fejljagt: Slangespil

Denne notebook indeholder flere små bugs, som du skal finde og rette. Hver bug du løser vil få spillet til at virke bedre!

**Hint:** Læs fejlmeddelelserne og tænk på hvad der kunne være galt.

Kør alle celler for at starte. Hvis spillet crasher eller misbehaver, jeg et fejljagt-kommentar for at finde bugget.

```python
import turtle
import time
import random
```

```python
# FEJLJAGT: Kontroller typen af variablerne
forsinkelse = "0.1"
point = 0
bedste_point = 0
spil_bredde = 600
spil_højde = 600
vindue_bredde = 700
vindue_højde = 700
```

```python
# FEJLJAGT: Kontroller disse knapper!
knap_op, knap_ned, knap_venstre, knap_højre = "Up", "Down", "Venstre", "Højre"
retning = "stop"
```

```python
skærm_titel = "Slangespil"
slange_farve = "green"
slange_hoved_form = "square"
slange_segment_form = "square"
mad_farve = "red"
mad_form = "circle"
baggrunds_farve = "black"
point_farve = "white"
point_skrifttype = ("Arial", 24, "normal")
```

```python
# FEJLJAGT: Kontroller disse variabler!
# Beregninger med forskellige matematiske operatorer:
x_min = -spil_bredde // 2  # // heltalsdivision
x_max = spil_bredde // 0
y_min = -spil_højde // 2  # - minus/negation
y_max = spil_højde // 2
point_x = 0
point_y = spil_højde // 2 - point_skrifttype[1] - 10  # - minus for at trække fra

# Flere eksempler på matematiske operatorer:
samlet_areal = spil_bredde * spil_højde  # * gange
halv_bredde = spil_bredde / 2  # / almindelig division (giver float)
dobbelt_forsinkelse = forsinkelse + forsinkelse  # + plus for at lægge sammen
margin = 10 + 5  # + plus
```

```python
# FEJLJAGT: Flytter slangen sig i den rigtige retning?
def flyt(hoved, retning):
    if retning == "op":
        y = hoved.ycor()
        hoved.sety(y + 20)
    if retning == "ned":
        y = hoved.ycor()
        hoved.sety(y - 20)
    if retning == "venstre":
        x = hoved.xcor()
        hoved.setx(x - 20)
    if retning =="højre":
        x = hoved.xcor()
        hoved.setx(x - 20)


def gå_op():
    global retning
    if retning != "ned":
        retning = "op"


def gå_ned():
    global retning
    if retning != "op":
        retning = "ned"


def gå_venstre():
    global retning
    if retning != "højre":
        retning = "venstre"


def gå_højre():
    global retning
    if retning != "venstre":
        retning = "højre"
```

```python
vindue = turtle.Screen()
vindue.title(skærm_titel)
vindue.bgcolor(baggrunds_farve)
vindue.setup(width=vindue_bredde, height=vindue_højde)
vindue.tracer(0)
```

```python
# FEJLJAGT: STOR fejl! Men svær at spotte
# prøv at spille spillet og se hvad der sker når du spiser mad.

hoved = turtle.Turtle()
hoved.speed(0)
hoved.shape(slange_hoved_form)
hoved.color(slange_farve)
hoved.penup()
hoved.goto(0, 0)
retning = "stop"
slange = (hoved,) 
```

```python
mad = turtle.Turtle()
mad.speed(0)
mad.shape(mad_form)
mad.color(mad_farve)
mad.penup()
mad.goto(0, 100)
```

```python
pen = turtle.Turtle()
pen.speed(0)
pen.shape("square")
pen.color(point_farve)
pen.penup()
pen.hideturtle()
pen.goto(point_x, point_y)
pen.write(f"Point: {point}  Bedste: {bedste_point}", align="center", font=point_skrifttype)
```

```python
vindue.listen()
vindue.onkeypress(gå_op, knap_op)
vindue.onkeypress(gå_ned, knap_ned)
vindue.onkeypress(gå_venstre, knap_venstre)
vindue.onkeypress(gå_højre, knap_højre)
```

```python
# FEJLJAGT: Er det den rigtige logik som er blevet brugt i kollisionen?
# FEJLJAGT: Kan slangen spise maden? hint: tjek afstanden mellem hoved og mad
# FEJLJAGT: MÅ MAN GODT BEDE OM AT FÅ DE POINT SOM MAN HAR FORTJENT? hint: tilføj point når man spiser mad
while True:
    vindue.update()

    grænse_kollision = (
        hoved.xcor() > x_max
        or hoved.xcor() < x_min
        or hoved.ycor() > y_max
        or hoved.ycor() < y_min
    )
    slange_kollision = any(
        segment.distance(hoved) < 20 for segment in slange[1:]
    )
    if grænse_kollision and slange_kollision:
        retning = "stop"
        for segment in slange:
            segment.shape("turtle")
            vindue.update()
            time.sleep(0.5)
        [segment.reset() for segment in slange[1:]]

        hoved.goto(0, 0)
        hoved.shape(slange_hoved_form)
        slange = [hoved]

        point = 0

        pen.clear()
        pen.write(
            f"Point: {point}  Bedste: {bedste_point}",
            align="center",
            font=point_skrifttype,
        )

    # FEJLJAGT #5: Hvornår skal slangen spise maden?
    if hoved.distance(mad) < 1:  
        nyt_segment = turtle.Turtle()
        nyt_segment.speed(0)
        nyt_segment.shape(slange_segment_form)
        nyt_segment.color(slange_farve)
        nyt_segment.penup()
        nyt_segment.goto(mad.xcor(), mad.ycor())
        slange.append(nyt_segment)

        x = random.randint(x_min, x_max)
        y = random.randint(y_min, y_max)
        mad.goto(x, y)

        point += 0

        if point > bedste_point:
            bedste_point = point
            pen.clear()
            pen.write(
                f"Point: {point}  Bedste: {bedste_point}",
                align="center",
                font=point_skrifttype,
            )

    for i in range(len(slange) - 1, 0, -1):
        if i > 0:
            x = slange[i - 1].xcor()
            y = slange[i - 1].ycor()
        slange[i].goto(x, y)

    flyt(hoved, retning)

    time.sleep(0.2)
```
